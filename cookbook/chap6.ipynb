```python
##ch5. correlation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
e = pd.read_excel('data/ecm_result4.4.xlsx')
```

```python
e.drop(columns=e.columns[0:5],inplace=True)
```

```python
corr = e.corr()
```

```python
corr[['kin','elementary','grdp','seoul_dist','ep','ur','er']]
```

```python

fig, axes = plt.subplots(1,2, sharey=True)
sns.regplot(x=e.ur, y=e.er, ax=axes[0])
sns.regplot(x=e.grdp, y=e.er, ax=axes[1])
axes[0].set_xlabel("실업률")
axes[0].set_ylabel("er")
axes[1].set_xlabel("grdp")
axes[1].set_ylabel("")
plt.tight_layout()
fig.subplots_adjust(top=0.92)
plt.show()
```

```python
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='coolwarm')
plt.tight_layout()
plt.show()
```

```python
#야간에 수면을 6시간 미만으로 취하는 이유에 따라 조사 응답자 범주화
def getsleepdeprivedreason(row):
    reason = "Unknown"
    if (row.nightlyhrssleep >= 6):
        reason = "Not sleep Deprived"
    elif (row.nightlyhrssleep > 0):
        if (row.weeksworked16+row.weeksworked17 < 80):
            if (row.childathome>2):
                reason = "child rearing"
            else:
                reason = "Others"
        else:
            if(row.wageincome>=62000 or row.highestgradecompleted>=16):
                reason = "Work pressure"
            else:
                reason = "Income pressure"
    else:
        reason = "Unknown"
    
    return reason
```

```python
nls97["SLEEP_REASON"] = nls97.apply(getsleepdeprivedreason, axis=1)
```

```python
nls97.SLEEP_REASON = nls97.SLEEP_REASON.astype('category')
nls97.SLEEP_REASON.value_counts()
```
